# Vorhersage von Schlafstörungen
### Datensatz:
**Sleep Health and Lifestyle Dataset**

# Problembeschreibung

Die frühzeitige und zuverlässige Erkennung von Krankheiten ist ein wichtiges Ziel in der Medizin. Die Identifizierung von Risikofaktoren für gewisse Krankeheiten ist für die Prävention essentiell.

Ich habe mich im Rahmen dieses Projekts für einen Datensatz entschieden, der die Zusammenhänge verschiedenster Gesundheitsdaten und Lebensweisen in Bezug auf Schlafstörungen thematisiert.
Heutzutage, und vor allem in unserem Alter, wird die Signifikanz von Schlaf häufig unterbewertet. Arbeit, Uni, Freizeit, Stress; all das führt dazu, dass man zu wenig schläft. Aber auch ungesunde Ernährung oder zu wenig Bewegung können zu ernsthaften Gesundheitsproblemen führen.
Dieses Projekt zielt darauf ab, genau diese Faktoren zu untersuchen und ein Modell zu trainieren, welches eine Vorhersage trifft, ob eine Person von einer Schlafstörung betroffen ist oder nicht. Mithilfe dieses Modells soll auch vorhergesagt werden, welche Faktoren für ein besonders hohes Risiko sorgen, an einer Schlafstörung zu erkranken.

## 2. Train/-Test-Split

Bevor das Modell trainiert wird, muss der Datensatz gesplittet werden, damit die Daten mit denen getestet wird das Training nicht beeinflussen.

```python
X = df.drop(columns=["Sleep Disorder"])

y = df["Sleep Disorder"]

X.shape, y.head()
```

```python
X_train_uncleaned, X_test_uncleaned, y_train_uncleaned, y_test_uncleaned = train_test_split(X, y, random_state=28, test_size=0.2)
X_train_uncleaned.shape, y_train_uncleaned.shape, X_test_uncleaned.shape, y_test_uncleaned.shape
```

## 3. Datenbereinigung

```python
X_train, X_test = X_train_uncleaned.copy(), X_test_uncleaned.copy()

X_train["BMI Category"] = X_train["BMI Category"].replace("Normal Weight", "Normal")
X_test["BMI Category"] = X_test["BMI Category"].replace("Normal Weight", "Normal")

# neu plotten
sns.countplot(data=X_train, x="BMI Category", palette="Set2", edgecolor="black", stat="percent", hue="BMI Category")

plt.show()
```

Wie man im Diagramm sehen kann, sind jetzt nur noch die drei Kategorien vorhanden. 

Außerdem wurde eine Kopie des Datensatzes erstellt, mit dem im weiteren gearbeitet wird.

Als nächstes wird der Wert für die gesunden Personen so verändert, dass sie im Diagramm auftauchen. 
Momentan sind die Werte für gesunde Personen, in der Kategorie "Sleep Disorer" NaN (Not a Number), da sie im ursprünglichen Datensatz "None" waren. 

Sie werden jetzt so ersetzt, dass der Wert "Healthy" heißt.
Davor wird noch einmal geprüft, welche Kategorien es gibt.

```python
# prüfen, welche Kategorien es gibt
y_train_uncleaned.unique()
```

```python
y_train = y_train_uncleaned.fillna("Healthy")
y_test = y_test_uncleaned.fillna("Healthy")

# erneute Prüfung
y_train.unique()
```

```python
# neu plotten
sns.countplot(data=y_train, palette="Set2", edgecolor="black", stat="percent")

plt.show()
```

Jetzt soll das Feature "Blutdruck" so verändert werden, dass damit gearbeitet werden kann. Dafür werden zwei neue Features erstellt, "Systolic" und "Diastolic", also der obere und untere Wert. Beide werden als integer gespeichert.
Zuletzt wird das alte Feature gelöscht, da es keinerlei neue Information bereitstellt.

```python

# expand = True: sorgt dafür, dass die beiden Werte in neue Spalten geschrieben werden und nicht in eine als Liste ["120", "80"]
X_train[["Systolic", "Diastolic"]] = X_train["Blood Pressure"].str.split("/", expand=True)
X_test[["Systolic", "Diastolic"]] = X_test["Blood Pressure"].str.split("/", expand=True)

# der aktuelle Datentyp
print(X_train[["Systolic", "Diastolic"]].dtypes)

# da Datentyp: str -> Umwandeln in int
X_train["Systolic"], X_train["Diastolic"] = X_train["Systolic"].astype(int), X_train["Diastolic"].astype(int)
X_test["Systolic"], X_test["Diastolic"] = X_test["Systolic"].astype(int), X_test["Diastolic"].astype(int)

X_train = X_train.drop(columns=["Blood Pressure"])
X_test = X_test.drop(columns=["Blood Pressure"])

X_train.head()
```

## 5. Encoding, Feature Engineering & Feature Selection

#### Binäry encoding für Binäre Features (Gender)

```python
X_train["Gender"] = X_train["Gender"].map({"Male": 0, "Female": 1})
X_test["Gender"] = X_test_uncleaned["Gender"].map({"Male": 0, "Female": 1})
X_train.head()
```

#### One-Hot Encoding für die Kategorialen Features (Occupation, BMI Category)

```python
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False).set_output(transform="pandas")
```

Encoder auf Trainingsdaten anwenden:

```python
transformed_train_data = encoder.fit_transform(X_train[["Occupation", "BMI Category"]])

transformed_train_data.head()
```

Und auf die Testdaten:

(Wichtig: encoder.transform, nicht encoder.fit_transform, da der fit bereits auf dem Trainingsdaten gemacht wurde)

```python
transformed_test_data = encoder.transform(X_test[["Occupation", "BMI Category"]])
transformed_test_data.head()
```

Die Datensätze jeweils zu einem finalen Datensatz zusammenfassen:

```python
X_train_final = pd.concat([X_train.drop(columns=["Occupation", "BMI Category"]), transformed_train_data], axis=1)

X_test_final = pd.concat([X_test.drop(columns=["Occupation", "BMI Category"]), transformed_test_data], axis=1)
X_test_final.head()
```

#### Label encoding für Zielvariable

```python
target_mapping = {"Healthy": 0, "Insomnia": 1, "Sleep Apnea": 2}

y_train = y_train.map(target_mapping)
y_test = y_test.map(target_mapping)
```

#### Feature Engineering

Aus den beiden verschiedenen aber dennoch eng miteinander verbundenen Features "Quality of Sleep" und "Sleep Duartion" lässt sich ein Schlafindex berechnen, der beide Features vereint. Für diesen gilt:
$$
\text{Sleepindex} = \frac{\text{Quality of Sleep}}{\text{Sleep Duaration}}
$$

Ein hoher Wert bedeutet kurzer aber erholsamer Schlaf und ein niedriger Wert bedeutet langer aber wenig erholsamer Schlaf.

```python
X_train_final["Sleep Index"] = (X_train_final["Quality of Sleep"] / X_train_final["Sleep Duration"]).round(2)
X_test_final["Sleep Index"] = (X_test_final["Quality of Sleep"] / X_test_final["Sleep Duration"]).round(2)

X_train_final.head()
```

#### Feature Selection

Im nächsten Schritt sollen alle Features, die keinen oder kaum Informationen liefern oder das Modell verwirren könnten, entfernt werden.

Folgende Features werden entfernt:
- Person ID: bringt keine Informationen in Bezug auf eine Schlafstörung
- Sleep Duration & Quality of Sleep: wurde durch den Schlafindex ersetzt
- Daily Steps: sehr ähnlich zu "Physical Activity Level" -> hohe Korrelation (siehe Korrelationsmatrix)
- Diastolic: korreliert hochgradig mit Systolic, deshalb kann dieses Feature weggelassen werden

```python
X_train_final = X_train_final.drop(columns=["Person ID", "Sleep Duration", "Quality of Sleep", "Daily Steps", "Diastolic"])

X_train_final.head()
```

```python
X_test_final = X_test_final.drop(columns=["Person ID", "Sleep Duration", "Quality of Sleep", "Daily Steps", "Diastolic"])

X_test_final.head()
```

## 6. Modelierung

Für die Vorhersage wird ein Random Forest trainiert. 
Dieser ist für Klassifikation gut geeiget, robust und benötigt keine Skalierung.

```python
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=11)

random_forest_model.fit(X_train_final, y_train)
```

```python
prediction = random_forest_model.predict(X_test_final)
```

## 7. Auswertung

```python
acc = accuracy_score(y_test, prediction)
prec = precision_score(y_test, prediction, average="weighted")
rec = recall_score(y_test, prediction, average="weighted")
f1 = f1_score(y_test, prediction, average="weighted")
f1_macro = f1_score(y_test, prediction, average="macro")
prec_macro = precision_score(y_test, prediction, average="macro")
rec_macro = recall_score(y_test, prediction, average="macro")

data = {
    "Metrik": ["Precision", "Recall", "F1-Score"],
    "Weighted": [prec, rec, f1],
    "Macro": [prec_macro, rec_macro, f1_macro]
}
res_df = pd.DataFrame(data=data)
res_df.set_index("Metrik", inplace=True)

display(res_df)
print(f"Accuracy: {acc:.2%}")
```

Für den weighted Score gilt, dass alle Klassen abhängig von ihrer Häufigkeit Einfluss nehmen. Macht eine Klasse einen großen Teil des Datensatzes aus, so macht diese auch einen entsprechend großen Teil des Scores aus.

Für den Macro Score gilt, dass der Durchschnitt einfach aus allen Klassen berechnet wird. Ist das Ergebnis in einer kleinen Klasse schlechter, hat das einen deutlich größeren Einfluss auf den Score. Da die Klassenverteilung in diesem Datensatz ungleich ist, ist der Macro Score besonders interessant. Denn er zeigt, ob die Vorhersage der kleinen Klassen auch zuverlässig funktioniert oder ob das Modell einfach immer "Gesund" (große Klasse) vorhersagt. 

#### Auf Overfitting prüfen

```python
y_train_pred = random_forest_model.predict(X_train_final)
train_acc = accuracy_score(y_train, y_train_pred)

print(f"{train_acc:.2%}")
```

Die Ergebnisse sind auf den Trainingsdatensatz leicht besser, der Unterschied ist aber nicht so hoch, dass es sich um Overfitting handelt.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6)) 
conf_matrix = confusion_matrix(y_test, prediction)

confusion_matrix_percent = confusion_matrix(y_test, prediction, normalize='true')

# absolute Zahlen
sns.heatmap(conf_matrix, ax=axes[0], annot=True, fmt="d", cmap="Blues", 
            xticklabels=target_mapping.keys(), 
            yticklabels=target_mapping.keys())
axes[0].set_title("Confusion Matrix (Absolut)")
axes[0].set_xlabel("Vorhergesagt")
axes[0].set_ylabel("Tatsächlich")

# relative Zahlen
sns.heatmap(confusion_matrix_percent, ax=axes[1], annot=True, fmt=".2%", cmap="Blues", 
            xticklabels=target_mapping.keys(),
            yticklabels=target_mapping.keys())
axes[1].set_title("Confusion Matrix (Relativ in %)")
axes[1].set_xlabel("Vorhergesagt")
axes[1].set_ylabel("Tatsächlich")

plt.tight_layout()
plt.show()
```

Anhand der Konfusionsmatrix lässt sich ablesen, dass das Modell die gesunden Personen sehr treffsicher vorhersagen kann. Schlafstörungen werden eher vorsichtig vorhergesagt, was zu einer erhöhten False Negative Rate führt. Das kann man in der Matrix in den rechten unteren Spalten ablesen: Insomnie wurde in 15.79% der Fällen fälschlicherweise als Gesund vorhergesagt und Schlaf Apnoe in  11,11% der Fälle.

```python
# Rangliste der Wichtigkeit erstellen
importances = random_forest_model.feature_importances_
feature_names = X_train_final.columns
feature_importance_df = pd.DataFrame({"Feature": feature_names, "Importance": importances})
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

# Visualisierung
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=feature_importance_df, palette="magma", hue="Feature")
plt.title("Feature Importance")
plt.show()
```

## 8. kurze Zusammenfassung

Die Auswertung der Ergebnisse zeigt, dass das Modell die Schlafstörungen zuverlässig vorhersagen kann. Die Feature Improtance hat ergeben, dass der Blutdruck, sowie der BMI und das Alter einen starken Zusammenhang mit Schlafstörungen hat. 

Ingesamt wurde das Ziel des Projekts erreicht. In Zukunft könnte man das Modell mit einem größeren Datensatz trainieren, um die Generalisierungfähigkeit zu verbessern oder durch Over- oder Undersampling die Klassenverteilung beeinflussen, um die Vorhersage auf den kleinen Klassen noch weiter zu verbessern.
